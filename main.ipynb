```python
import requests
from bs4 import BeautifulSoup

# need this to conform with wikipedia settings
HEADERS = {
    "User-Agent": "cisd-43-sentiment-analysis/0.1 (educational project; contact: brandoncabrera707@gmail.com)",
    "Accept-Language": "en-US,en;q=0.9",
}
```

```python
url = 'https://en.wikipedia.org/wiki/Biostatistics'
response = requests.get(url, headers=HEADERS, timeout=30)
response.raise_for_status()
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify()) # information about page layout
```

# Title 

```python
title = soup.title.text
print(title)
```

# Introduction Paragraph

```python
intro_text = soup.find('p').text
print(intro_text)
```

# History 

```python
history = soup.find(class_ = 'mw-heading mw-heading3') # div that contains Biostatistics and genetics header
history_text = ""

for sibling in history.find_next_siblings():
  if sibling.text != "Research planning[edit]": #History Section ends when this one starts
    history_text += sibling.text
  else:
    break

print(history_text)
```

```python
main_content = soup.find(id = "Research_planning")
main_content = main_content.find_parent() # div that contains research planning header
main_content_text = ""
for sibling in main_content.find_next_siblings():
  if sibling.text != "References[edit]":
    main_content_text += sibling.text
  else:
    break
print(main_content_text)
```

# References 

```python
for link in soup.find_all('a'): # found in beautifulsoup's documentation and is used to find all refrences 
    print(link.get('href'))
```

# Sentiment Analysis


```python
from textblob import TextBlob

intro_sentences = intro_text.split(".")
history_sentences = history_text.split(".")
main_content_sentences = main_content_text.split(".")
```

```python
def classify_polarity(sentence):
  polarity = TextBlob(sentence).sentiment.polarity
  if polarity > 0.5:
      return 'Positive'
  elif polarity < -0.5:
      return 'Negative'
  else:
      return 'Neutral'
    
def classify_subjectivity(sentence):
  subjectivity = TextBlob(sentence).sentiment.subjectivity
  if subjectivity > 0.90:
    return "Highly Subjective"
  elif subjectivity > 0.60:
    return "Moderately Subjective"
  elif subjectivity > 0.30:
    return "Mild subjectivity"
  else:
    return "Barely any subjectivity"
```

```python
for sentence in intro_sentences:
  polarity = classify_polarity(sentence)
  subjectivity = classify_subjectivity(sentence)
  print(f"Sentence: {sentence} \nPolarity: {polarity} \nSubjectivity: {subjectivity} \n ")
```

```python
for sentence in history_sentences:
  polarity = classify_polarity(sentence)
  subjectivity = classify_subjectivity(sentence)
  print(f"Sentence: {sentence} \nPolarity: {polarity} \nSubjectivity: {subjectivity} \n ")
```

```python
for sentence in main_content_sentences:
  polarity = classify_polarity(sentence)
  subjectivity = classify_subjectivity(sentence)
  print(f"Sentence: {sentence} \nPolarity: {polarity} \nSubjectivity: {subjectivity} \n ")
```
